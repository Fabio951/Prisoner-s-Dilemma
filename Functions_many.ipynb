## Challenge normal

```python
def result(data):
    T = 3
    R = 2
    P = 1
    S = 0
    outcomes = np.array([[[P,P],[T,S]], [[S,T], [R,R]]])
    return outcomes[data[-1][0],data[-1][1]]
    
    
```

```python
def update_res(data,pris_a,pris_b):
    pris_a.result += result(data)[0]
    pris_b.result += result(data)[1]
```

```python
def challenge(pris1,pris2,n_times):
    pris1.contestant = 0
    pris2.contestant = 1
    data = np.zeros((0,2), dtype=int)
    for i in np.arange(n_times):
        data = np.append(data,[[pris1.choice(data),pris2.choice(data)]],axis=0)
        update_res(data,pris1,pris2)
    return()
```

```python
def challenge_print(pris1,pris2,n_times):
    pris1.contestant = 0
    pris2.contestant = 1
    data = np.zeros((0,2), dtype=int)
    print(pris1.type,' vs ', pris2.type)
    for i in np.arange(n_times):
        data = np.append(data,[[pris1.choice(data),pris2.choice(data)]],axis=0)
        update_res(data,pris1,pris2)
        #print(pris1.result,pris2.result)
    print(data)
    return()
```

```python
def evaluate_gen_all_vs_all(gen,display_bar=False, n_turns=100,step=10):
    reset(gen)
    display_bar_plot(display_bar=display_bar,initial=True)
    for i in range(len(gen)-1):
        display_bar_plot(display_bar=display_bar,later=True,step=step,n_gen=len(gen))
        for j in range(i+1,len(gen)):
            if i==j:
                break
            challenge_with_det(gen[i],gen[j],n_turns)
    display_bar_plot(display_bar=display_bar,close=True)      
```

```python
def evaluate_gen_one_vs_one(gen, n_turns=100):
    reset(gen)
    opponents = np.random.choice(len(gen),(int(len(gen)/2),2),replace=False)
    for couple in opponents:
        challenge_with_det(gen[couple[0]],gen[couple[1]],n_turns)    
```

```python
def evaluate_gen_n_vs_n(gen, n_turns=100,n_per_group=10):
    reset(gen)
    opponents = np.random.choice(len(gen),(int(len(gen)/n_per_group),n_per_group),replace=False)
    for group in opponents:
        for i in range(n_per_group-1):
            for j in range(i+1,n_per_group):
                challenge_with_det(gen[group[i]],gen[group[j]],n_turns)
```

```python
def evaluate_gen_all_vs_n(gen,n_turns=100,n_challengers=10):
    challengers = np.random.choice(len(gen),n_challengers,replace=False)
    challengers = [gen[i] for i in challengers]
    for pris1 in gen:
        for pris2 in challengers:
            challenge_with_det_avn(pris1,pris2,n_turns)
```

```python
def random_gen(n=100,all_selfish=False,n_turns=100,selection_classes=[False]):
    if all_selfish==False:
        if selection_classes[0]==True:
            if selection_classes[1]==1:
                gen = np.random.choice([0,1,2,3,4],n).tolist()
                gen = translate_gen(gen,n_turns)
            else:
                classes_used = [i-2 for i in range(2,len(selection_classes)) if selection_classes[i]==1]
                gen = np.random.choice(classes_used,n).tolist()
                gen = translate_gen(gen,n_turns)
        else:
            gen = np.random.randint(0,20,n).tolist()
            gen = translate_gen(gen,n_turns)
    else:
        gen = (np.ones(n)*2).tolist()
        gen = translate_gen(gen,n_turns)
    return gen
```

```python
def translate_gen(gen,n):
    for i in range(len(gen)):
        if gen[i]==0:
            gen[i] = tft()
            continue
        if gen[i]==1:
            gen[i] = always_coop()
            continue
        if gen[i]==2:
            gen[i] = always_def()
            continue
        if gen[i]==3:
            gen[i] = mainly_nice(k=np.random.uniform(0.5,1))
            continue
        if gen[i]==4:
            gen[i] = mainly_bad(k=np.random.uniform(0,0.5))
            continue
        if gen[i]==5:
            gen[i] = random()
            continue
        if gen[i]==6:
            gen[i] = ntft()
            continue
        if gen[i]==7:
            gen[i] = angry()
            continue
        if gen[i]==8:
            gen[i] = win_stay(initial=np.random.choice(2))
            continue
        if gen[i]==9:
            gen[i] = tft_forgive(initial=np.random.choice(2),forgive=np.random.choice(np.arange(4)+2,p=(2/3,1/6,1/12,1/12)))
            continue
        if gen[i]==10:
            gen[i] = tft_punish(initial=np.random.choice(2),apologize=np.random.choice(np.arange(4),p=(1/3,1/3,1/6,1/6)))
            continue
        if gen[i]==11:
            gen[i] = majority(initial=np.random.choice(2),n_moves=np.random.choice(np.arange(2,n)))
            continue
        if gen[i]==12:
            gen[i] = methodical(initial=np.random.choice(2),probs=np.random.uniform(size=4))
            continue
        if gen[i]==13:
            gen[i] = changer(in_prob=np.random.uniform(),step=np.random.uniform()/5)
            continue
        if gen[i]==14:
            gen[i] = r_spq(s=np.random.uniform(),p=np.random.uniform(),q=np.random.uniform())
            continue
        if gen[i]==15:
            gen[i] = equalizer(initial=np.random.choice(2))
            continue
        if gen[i]==16:
            gen[i] = extort(initial=np.random.choice(2))
            continue
        if gen[i]==17:
            gen[i] = generous(initial=np.random.choice(2))
            continue
        if gen[i]==18:
            gen[i] = gang()
            continue
        if gen[i]==19:
            gen[i] = antitft()
            continue
    return gen
```

```python
def translate(pris,n):
    if (pris==0 or pris=='tft'):
        pris = tft()
    if (pris==1 or pris=='Good'):
        pris = always_coop()
    if (pris==2 or pris=='Bad'):
        pris = always_def()
    if (pris==3 or pris=='good'):
        pris = mainly_nice(k=np.random.uniform(0.5,1))
    if (pris==4 or pris=='bad'):
        pris = mainly_bad(k=np.random.uniform(0,0.5))
    if (pris==5 or pris=='random'):
        pris = random()
    if (pris==6 or pris=='ntft'):
        pris = ntft()
    if (pris==7 or pris=='angry'):
        pris = angry()
    if (pris==8 or pris=='winstay'):
        pris = win_stay(initial=np.random.choice(2))
    if (pris==9 or pris=='tftforgive'):
        pris = tft_forgive(initial=np.random.choice(2),forgive=np.random.choice(np.arange(4)+2,p=(2/3,1/6,1/12,1/12)))
    if (pris==10 or pris=='tftpunish'):
        pris = tft_punish(initial=np.random.choice(2),apologize=np.random.choice(np.arange(4),p=(1/3,1/3,1/6,1/6)))
    if (pris==11 or pris=='majority'):
        pris = majority(initial=np.random.choice(2),n_moves=np.random.choice(np.arange(2,n)))
    if (pris==12 or pris=='methodical'):
        pris = methodical(initial=np.random.choice(2),probs=np.random.uniform(size=4))
    if (pris==13 or pris=='changer'):
        pris = changer(in_prob=np.random.uniform(),step=np.random.uniform()/5)
    if (pris==14 or pris=='r_spq'):
        pris = r_spq(s=np.random.uniform(),p=np.random.uniform(),q=np.random.uniform())
    if (pris==15 or pris=='equalizer'):
        pris = equalizer(initial=np.random.choice(2))
    if (pris==16 or pris=='extort'):
        pris = extort(initial=np.random.choice(2))
    if (pris==17 or pris=='generous'):
        pris = generous(initial=np.random.choice(2))
    if (pris==18 or pris=='gang'):
        pris = gang()
    if (pris==19 or pris=='antitft'):
        pris = antitft()
    return pris
```

```python
def reset(gen):
    for i in range(len(gen)):
        gen[i].result = 0
```

```python
def make_chart(gen):
    chart = {'Prisoner type': [pris.type for pris in gen],
             'Result':[pris.result for pris in gen]}
    chart = pd.DataFrame(data=chart)
    chart = chart.sort_values(by=['Result'],ascending=False).reset_index(drop=True)
    chart.index +=1
    return chart
```

## Evolution and mutation

```python
def evolve(gen,offset=1,power=2.9,show=False,controlled_mutation=0.1,random_mutation=0.05,prob_gang=0.05,
           prob_antitft=0.05,n_rounds=100,selection_classes=[False]):
#offset sposta in alto la distribuzione in modo tale da non uccidere tutte quelle più basse,
#power amplifica il segnale in modo da rendere le prime più probabili
    from scipy.stats import halfnorm
    chart = make_chart(gen)
    result = np.array(chart.Result)
    range_results = (max(result)-min(result))
    if range_results!=0:
        pdf_evolution = ((result-min(result))/range_results)**power + offset/100
        pdf_evolution = pdf_evolution/np.sum(pdf_evolution)
    else:
        pdf_evolution = result/np.sum(result)
    selection = np.sort(np.random.choice(len(gen),int(len(gen)/2),p=pdf_evolution,replace=False).astype(int))
    if show==True:
        show_evolution(selection,gen)
    sorted_gen = sort_generation(gen)
    gen_new = [sorted_gen[i] for i in selection]
    gen_new = generate_mutations(gen_new,n_rounds=n_rounds,controlled_mutation=controlled_mutation,
                                 random_mutation=random_mutation,prob_gang=prob_gang,prob_antitft=prob_antitft,
                                 selection_classes=selection_classes)
    reset(gen_new)
    return(gen_new)
```

```python
def generate_mutations(gen_new,n_rounds,controlled_mutation,random_mutation,prob_gang,prob_antitft,
                       selection_classes=[False]):
    gen_add = []
    count_tft = [1 for pris in gen_new if (pris.class_n==0 or pris.class_n==6)]
    count_tft = np.sum(count_tft)/len(gen_new)
    if selection_classes[0]==True:
        selected_classes = True
        if selection_classes[1]==1:
            possible_classes = [0,1,2,3,4]
        else:
            possible_classes = [i-2 for i in range(2,len(selection_classes)) if selection_classes[i]==1]
    else:
        selected_classes = False
    
    for prisoner in gen_new:
        if selected_classes:
            accepted = False
            while not accepted:
                children = prisoner.mutate(controlled_mutation=controlled_mutation,random_mutation=random_mutation,
                                           prob_gang=prob_gang,prob_antitft=prob_antitft,count_tft=count_tft,
                                           n=n_rounds,possible_classes=possible_classes)
                for pris in children:
                    if not pris.class_n in possible_classes:
                        accepted = False
                        break
                    accepted = True
        else:
            children = prisoner.mutate(controlled_mutation=controlled_mutation,random_mutation=random_mutation,
                                        prob_gang=prob_gang,prob_antitft=prob_antitft,count_tft=count_tft,
                                        n=n_rounds)
        gen_add += children
    return(gen_add)
```

```python
def show_evolution(selection,gen):
    dimension=np.sqrt(len(gen)).astype(int)
    if np.abs(dimension**2-len(gen))>.5:
        dimension +=1
    selection = np.array(selection)
    mask = np.zeros(len(gen)).astype(int)
    mask[selection] = 1
    x = np.arange(len(gen))%dimension + 1
    y = dimension - ((np.arange(len(gen))/dimension)%dimension).astype(int)
    while y[-1]!=1:
        y -= 1 
    fig,ax = plt.subplots(figsize=(9,9))
    ax.scatter(x,y,c=mask,s=20,cmap='prism')
    ax.xaxis.set_ticks_position('none') 
    ax.set_xticklabels('')
    ax.set_yticks(range(1,max(y)+1))
    ax.set_yticklabels([str((i-1)*max(x)+1) + '-' + str((i)*max(x)) for i in range(max(y)+1)][::-1])
    plt.show()
```

```python
def second_element(elem):
    return elem[1]

def sort_generation(gen):
    results = [pris.result for pris in gen]
    gen_sorted = [[i,j] for i,j in zip(gen,results)]
    gen_sorted.sort(key=second_element,reverse=True)
    gen_sorted = [dat[0] for dat in gen_sorted]
    return gen_sorted
        
```
